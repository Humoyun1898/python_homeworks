1.Group the data by the Category column and calculate the following aggregate statistics for each category:
Total quantity sold.
Average price per unit.
Maximum quantity sold in a single transaction.

```python
import pandas as pd

sales = pd.read_csv('sales_data.csv')
sales
sales.groupby('Category').agg({'Quantity':['sum','max'],'Price':'mean'})

```

2.Identify the top-selling product in each category based on the total quantity sold.

```python
cat_prod_quan_sum = sales.groupby(['Category','Product']).agg({'Quantity':'sum'}).reset_index()
cat_prod_quan_sum.sort_values(['Category','Quantity'], ascending=False)
top_products = cat_prod_quan_sum.groupby('Category').first().reset_index()
top_products
```

3. Find the date on which the highest total sales (quantity * price) occurred.


```python
sales['Revenue'] = sales['Quantity'] * sales['Price']
sales_per_date = sales.groupby('Date').agg({'Revenue':'sum'}).reset_index()
best_date = sales_per_date.sort_values('Revenue', ascending=False)
best_date.head(1)
```

1.Group the data by CustomerID and filter out customers who have made less than 20 orders.

```python
orders = pd.read_csv('customer_orders.csv')
number_of_orders_per_customer = orders.groupby('CustomerID').agg({'OrderID':'count'}).reset_index()
filter = number_of_orders_per_customer['OrderID'] < 20
number_of_orders_per_customer[filter]
```

2.Identify customers who have ordered products with an average price per unit greater than $120.

```python
avg_price_per_unit = orders.groupby('Product').agg({'Price':'mean'})
filter = avg_price_per_unit['Price'] > 120
avg_price_per_unit[filter]
```

3.Find the total quantity and total price for each product ordered, and filter out products that have a total quantity less than 5 units.


```python
grouped = orders.groupby('Product').agg({'Quantity':'sum','Price':'sum'})
filter = grouped['Quantity'] < 5
grouped[filter]
```
